```python
import requests
import json
import pandas as pd
```

```python
url =   "https://macrostrat.org/api/v2/columns?project_id=10&format=geojson_bare&status_code=in%20process"
```

```python
res = requests.get(url)
data = res.json()
```

```python
features = data['features']

first = features[0]
params = {}
params['location'] = json.dumps(first['geometry'])
params['project_id'] = first['properties']['project_id']
params['col_group'] = first['properties']['col_group']
params['col_name'] = first['properties']['col_name']
params['col_id'] = first['properties']['col_id']
```
